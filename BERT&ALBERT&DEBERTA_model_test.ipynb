```python
import torch
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
x = torch.arange(0, 128).unsqueeze(0).to(device)
x
```

```python
from nano_bert.model import DEBertAMix3
deberta = DEBertAMix3(
    vocab_size=4000, 
    n_layers=2,
    n_heads=1,
    max_seq_len=128,
    n_classes=2,
    n_embed = 4
).to('cuda')
deberta
```

```python
deberta(x)
```

```python
from nano_bert.model import ALBertMix3
albert = ALBertMix3(
    vocab_size=4000, 
    n_layers=2,
    n_heads=1,
    max_seq_len=128,
    n_classes=2,
    n_embed = 4
).to('cuda')
albert
```

```python
albert(x)
```

```python
from nano_bert.model import BertMix3
bert = BertMix3(
    vocab_size=4000, 
    n_layers=2,
    n_heads=1,
    max_seq_len=128,
    n_classes=2,
    n_embed = 4
).to('cuda')
bert
```
